```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_csv = pd.read_csv("/content/Facebook Friends List.csv")
df_excel = pd.read_excel("/content/Facebook Friends.xlsx")

df_txt = pd.read_csv("/content/Facebook Friends.txt", encoding='utf-16', sep='\t')

with open("/content/Facebook Friends.txt", 'r', encoding='utf-16') as f:
    print("First 5 lines of TXT file:")
    print(''.join(f.readlines()[:5]))

print("CSV Shape:", df_csv.shape)
print("Excel Shape:", df_excel.shape)
print("TXT Shape:", df_txt.shape)

print("\nCSV Data Types:\n", df_csv.dtypes)
print("\nExcel Data Types:\n", df_excel.dtypes)
print("\nTXT Data Types:\n", df_txt.dtypes)

print("\nCSV Head:\n", df_csv.head())
print("\nExcel Head:\n", df_excel.head())
print("\nTXT Head:\n", df_txt.head())
```

```python
df = df_excel
df.columns
df.head()
df.tail()
df.info()
df.isnull().sum()
df.isnull().sum().sum()
```

```python
df.describe()
df.describe(include='all')
```

```python
df.info()
```

```python
df.hist(figsize=(15,12), bins=20, color = "yellow", edgecolor = "green")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.histplot(df["Age"], bins=15, kde=True, color = "green")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
print(df.columns)
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
import matplotlib.pyplot as plt


gender_counts = df['Gender'].value_counts()
labels = []
for key in gender_counts.index:
    if key == 1:
        labels.append('Male')
    elif key == 0:
        labels.append('Female')
    else:
        labels.append('Other/Unknown')

plt.figure()
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', colors=['cyan', 'orange', 'gray'])
plt.title("Gender Frequency")
plt.show()
```

```python
plt.figure()
sns.countplot(x="Relationship", data=df, color = "red")
plt.title("Relationship Status Distribution")
plt.xticks()
plt.show()
```

```python
plt.figure()
sns.boxplot(y=df["Friends"], color = "purple")
plt.title("Boxplot of Friends Count")
plt.show()
```
