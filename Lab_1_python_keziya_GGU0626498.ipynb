container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

```python
# Given values
#PROBLEM 1: Confusion Matrix
# Total predictions
# Accuracy
# Precision
# Recall
# F1 Score

TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN

# Metrics
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

# Output
print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```python
#PROBLEM 2: R²

#Find:
#SST
#R²
#Check if > 70%

# Given values
SSR = 420
SSE = 180

# Total sum of squares
SST = SSR + SSE

# R squared
R2 = SSR / SST

print("SST:", SST)
print("R2:", R2)

if R2 > 0.7:
    print("Model explains more than 70% variance")
else:
    print("Model explains less than 70% variance")
```

```python
#PROBLEM 3: Mean Squared Error
# Squared errors
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

# Mean Squared Error
mse = (e1 + e2 + e3 + e4 + e5) / 5

print("Mean Squared Error:", mse)
```

```python
#PROBLEM 4: Cook’s Distance
# Given values
ei2 = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Cook's Distance
cook = (ei2 / (p * MSE)) * (hii / (1 - hii) ** 2)

print("Cook's Distance:", cook)

threshold = 4 / n
print("Threshold:", threshold)

if cook > threshold:
    print("Observation is influential")
else:
    print("Observation is NOT influential")
```

```python
#PROBLEM 5: Regression Prediction
# Given values
a = 2   # intercept
b = 3   # slope
X = 5   # rainfall value

# Prediction
y = a + b * X

print("Predicted value:", y)
print("Slope shows change in output per unit input")
print("Intercept shows output when input is zero")
```
